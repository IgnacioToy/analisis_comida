```python
import pandas as pd
import numpy as np

from pathlib import Path
```

## 0. Configuración de rutas y configuración visual. 

```python
def cagar_archivo():
    ruta_archivo = Path.cwd().parent / "data" / "raw" / "dataset_comida.csv"
    try:
        df = pd.read_csv(ruta_archivo)
        print(f"""Carga exitosa del DataFrame, este se conforma de la siguiente manera:\n
                Filas del DataFrame: {df.shape[0]}\n
                Columnas del DataFrame: {df.shape[1]}
        """)

        return df

    except FileNotFoundError:
        print(f"    ERROR RUTA: No se encontro el archivo en {df.absolute()}")
        return None

    except Exception as e:
        print(f"    ERROR INESPERADO: Ocurrio un error inesperado en la carga del archivo: {e}")
        return None

df = cagar_archivo()
```

## 1. Normalización de texto (Standarization):

Para evitar duplicados por errores de escritura, conviert todas las categorías a formato título y elimino los espacion en blanco

```python
columnas_texto = ['city', 'order_time', 'cuisine', 'meal_type', 'mood', 'company']

for col in columnas_texto:
    df[col] = df[col].str.strip().str.title()
```

## 2. Transformación de variables Binarias (Encoding):

```python
# Mapeo de descuentos y otras binarias:
columnas_binarias = ['discount_applied', 'is_repeat_order', 'rainy_weather']

for col in columnas_binarias:
    df[col] = df[col].replace({'Yes': 1, 'No': 0})
```


Convierto variables categóricas de dos opciones (yes/no) en variables numéricas (1 / 0) para permitir cálculos estadísticos.

## 3. Ingenieria de caracteristicas (Feature Engineering)

```python
# Gasto total del cliente
df['total_amount'] = df['order_value'] + df['delivery_fee']

# Extraer la hora (asumiendo formato HH:MM)

mapping = {
    "Morning": 9,
    "Afternoon": 15,
    "Evening": 20,
    "Night": 23
}

df['order_hour'] = df['order_time'].map(mapping)
```
